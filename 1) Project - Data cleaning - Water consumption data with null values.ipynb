```python
# NOTE:- Dataset contains  missing values, So filling or dropping was required.
```

```python
# START CLEANING
```

```python
# Fill City (Categorical) with MODE 
```

```python
df["City"] = df["City"].fillna(df["City"].mode()[0])
```

```python
# now check null values in city column is filled with values or not
```

```python
df["City"].isnull()
```

```python
# count null values are there in city column
```

```python
df["City"].isnull().sum()
```

```python
#  Fill Numeric Columns with MEAN 
```

```python
df["Water_Consumption_Liters"]=df["Water_Consumption_Liters"].fillna(df["Water_Consumption_Liters"].mean())
```

```python
df["Water_Consumption_Liters"].isnull().sum()
```

```python
df["Household_Size"]=df["Household_Size"].fillna(df["Household_Size"].mean())
```

```python
df["Household_Size"].isnull().sum()
```

```python
df["Water_Bill_Amount"]=df["Water_Bill_Amount"].fillna(df["Water_Bill_Amount"].mean())
```

```python
df["Water_Bill_Amount"].isnull().sum()
```

```python
# Now verify dataset is cleaned or null 
```

```python
df.isnull().sum()
```
